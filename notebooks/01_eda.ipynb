# 01 — Exploratory Data Analysis
Telco customer churn dataset.

```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from src.data_loader import load_data
df = load_data()
df.head()
```

```python
df.info()
df.describe(include='all').T
```

```python
df['Churn'].value_counts(normalize=True).plot(kind='bar', title='Churn distribution');
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, col in zip(axes, ['Contract', 'InternetService', 'PaymentMethod']):
    pd.crosstab(df[col], df['Churn'], normalize='index').plot(kind='bar', stacked=True, ax=ax)
    ax.set_title(col)
plt.tight_layout()
```

```python
sns.kdeplot(data=df, x='tenure', hue='Churn', common_norm=False);
```
